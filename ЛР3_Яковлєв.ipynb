```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 200)
y1 = np.sin(x)
y2 = np.cos(x)

fig, ax = plt.subplots(figsize=(7, 3.5))
ax.plot(x, y1, linestyle='--', marker='s', markevery=10, label='sin(x)', color="green")
ax.plot(x, y2, linestyle='-', marker='o', markevery=10, label='cos(x)', color="purple")

ax.set_title("y = sin (x), y = cos (x)")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.grid(True, alpha=0.5)
ax.legend()
plt.show()
```

```python
np.random.seed(1)
x = np.random.randn(200)
y = 1.2 * x + np.random.randn(200) / 3

c = np.hypot(x, y)

fig, ax = plt.subplots(figsize=(7.5, 4))
sc = ax.scatter(x, y, c=c, s=65, edgecolor='b')
ax.set_title("Scatter: колір за величиною")
ax.set_xlabel("x")
ax.set_ylabel("y")
plt.colorbar(sc, ax=ax, label="шкала кольору")
ax.grid(True, alpha=0.5)

plt.show()
```

```python
cats = ["A1", "A2", "A3", "A4", "A5", "A6"]
vals = [21, 22, 3, 17, 8, 5]

fig, ax = plt.subplots(figsize=(10, 5))
bars = ax.bar(cats, vals)

ax.set_title("Bar chart із підписами")
ax.set_xlabel("Категорія")
ax.set_ylabel("Значення")
ax.grid(axis='y', alpha=0.6)

for b in bars:
    ax.text(b.get_x() + b.get_width()/2, b.get_height() + 0.1,
            f"{b.get_height():.0f}", ha="center", va="bottom")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3, 3, 200)
y = np.linspace(-3, 3, 200)
X, Y = np.meshgrid(x, y)
Z = np.sin(X**2 + Y**2)

fig, ax = plt.subplots(figsize=(6, 4))
cs = ax.contour(X, Y, Z, levels=15, cmap="viridis")
ax.clabel(cs, inline=True, fontsize=8)
ax.set_title("Contour plot Z(X,Y)")
ax.set_xlabel("X")
ax.set_ylabel("Y")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-4, 4, 100)
y = np.linspace(-4, 4, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(2*X**2 + 0.5*Y**2))

fig = plt.figure(figsize=(7, 5))
ax = fig.add_subplot(111, projection="3d")
surf = ax.plot_surface(X, Y, Z, cmap="plasma", edgecolor="none")
ax.set_title("3D Surface")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
fig.colorbar(surf, shrink=0.5, aspect=8)
plt.show()
```

```python
import matplotlib.gridspec as gridspec

np.random.seed(0)
t = np.linspace(0, 20, 300)
trend = 0.15*t + np.sin(t) + np.random.normal(0, 0.3, len(t))

x = np.random.randn(400)
y = 0.8*x + np.random.randn(400)/2

fig = plt.figure(figsize=(9, 6))
gs = gridspec.GridSpec(2, 2, height_ratios=[2, 1], width_ratios=[2, 2])

ax1 = fig.add_subplot(gs[0, :])
ax1.plot(t, trend, linewidth=2)
ax1.set_title("Trend line")
ax1.grid(True, alpha=0.3)

ax2 = fig.add_subplot(gs[1, 0])
ax2.hist(x, bins=30, alpha=0.7)
ax2.set_title("Histogram")
ax2.grid(True, alpha=0.3)

ax3 = fig.add_subplot(gs[1, 1])
ax3.scatter(x, y, alpha=0.6, edgecolor="k")
ax3.set_title("Scatter")
ax3.grid(True, alpha=0.3)

fig.subplots_adjust(top=0.9)
fig.suptitle("Dashboard з GridSpec", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x)

fig, ax = plt.subplots(figsize=(6, 3))
ax.plot(x, y, linewidth=2)
point, = ax.plot([], [], 'ro')
text = ax.text(0.5, 1.2, "", fontsize=10)

ax.set_title("Анімація точки на sin(x)")
ax.set_xlabel("x")
ax.set_ylabel("sin(x)")
ax.set_ylim(-1.5, 1.5)
ax.grid(True, alpha=0.3)

def init():
    point.set_data([], [])
    text.set_text("")
    return point, text

def update(frame):
    point.set_data([x[frame]], [y[frame]])
    text.set_text(f"x={x[frame]:.2f}, y={y[frame]:.2f}")
    return point, text

ani = FuncAnimation(fig, update, frames=len(x), init_func=init,
                    interval=30, blit=True)

HTML(ani.to_jshtml())


```
