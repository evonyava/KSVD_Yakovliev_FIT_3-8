```python
import seaborn as sns
```

```python
from google.colab import files

uploaded = files.upload()
```

```python
df = pd.read_csv("shopping_trends.csv", encoding='latin1')
df
```

```python
df.columns = df.columns.str.lower().str.replace(" ", "_")
```

```python
print(df.columns)
```

```python
sns.histplot(df['age'], bins=20, kde=True)
plt.title("Розподіл віку клієнтів")
plt.show()
```

```python
sns.countplot(y=df['item_purchased'], order=df['item_purchased'].value_counts().index)
plt.title("Найчастіше куплені товари")
plt.show()
```

```python
sns.scatterplot(data=df, x='age', y='purchase_amount_(usd)')
plt.title("Залежність віку від суми покупок")
plt.show()
```

```python
sns.histplot(df['review_rating'], bins=10, kde=True)
plt.title("Розподіл оцінок товарів")
plt.show()
```

```python
sns.countplot(x='gender', data=df)
plt.title("Розподіл покупок за статтю")
plt.show()
```

```python
sns.scatterplot(data=df, x='review_rating', y='purchase_amount_(usd)')
plt.title("Рейтинг vs сума покупок")
plt.show()
```

```python
import altair as alt
from vega_datasets import data
import plotly.express as px
```

```python
source = data.wheat()
chart1 = alt.Chart(source).mark_bar().encode(
    x='wheat:Q',
    y='year:O',
    text='wheat:Q'
).properties(title="Врожайність пшениці по роках")
chart1
```

```python
source = data.iowa_electricity()
chart2 = alt.Chart(source).mark_bar(opacity=0.7).encode(
    x='year:O',
    y='net_generation:Q',
    color='source:N'
).properties(title="Виробництво електроенергії в Айові")
chart2
```

```python
iris = px.data.iris()
fig = px.scatter(
    iris,
    x="sepal_width",
    y="sepal_length",
    size="petal_length",
    color="species",
    hover_data=["petal_width"]
)
fig.show()
```
